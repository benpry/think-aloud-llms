```python
# check there are no duplicate problems in the conditions
problems_set = []
for i, condition in enumerate(conditions):
    for problem in condition:
        if problem not in problems_set:
            problems_set.append(problem)
        else:
            print(f"Duplicate problem in condition {i+1}")
            break
```

```python
# get the average solve rate and AMT (s) for each condition
solve_rates = []
amts = []
for i, condition in enumerate(conditions):
    condition_solve_rate = []
    condition_amt = []
    for problem in condition:
        # find problem in problems dataframe
        problem_row = problems[problems["Puzzles"] == " ".join(map(str, problem))]
        condition_solve_rate.append(problem_row["Solve Rate"].values[0].item())
        condition_amt.append(problem_row["AMT (s)"].values[0].item())
    solve_rates.append(condition_solve_rate)
    amts.append(condition_amt)
```

```python
# run anova on the solve rates and AMT (s) and plot the results

import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Perform one-way ANOVA for Solve Rates
anova_solve = stats.f_oneway(*solve_rates)
print(
    f"ANOVA Result for Solve Rates: F={anova_solve.statistic}, p={anova_solve.pvalue}"
)

# Perform one-way ANOVA for AMT (s)
anova_amt = stats.f_oneway(*amts)
print(f"ANOVA Result for AMT (s): F={anova_amt.statistic}, p={anova_amt.pvalue}")

# Plotting Solve Rates
plt.figure(figsize=(12, 6))
sns.boxplot(data=solve_rates)
plt.title("ANOVA on Solve Rates Across Conditions")
plt.xlabel("Condition")
plt.ylabel("Solve Rate")
plt.show()

# Plotting AMT (s)
plt.figure(figsize=(12, 6))
sns.boxplot(data=amts)
plt.title("ANOVA on AMT (s) Across Conditions")
plt.xlabel("Condition")
plt.ylabel("AMT (s)")
plt.show()
```
